Retrieve the stock prices from Yahoo!

```python
start = datetime.datetime(2000, 1, 1)
end = datetime.datetime(2014, 1, 1)

stx = web.DataReader(dax_symbols.Symbol.values, data_source="yahoo", start=start, end=end)
```

Fill in missing closing prices by copying the previous value

```python
stx['Adj Close'] = stx.ix['Adj Close'].fillna(method='ffill', axis=0, limit=2)
```

Compute the daily returns.

$$ r(t) = log\left( \frac{p(t)}{p(t-1)} \right)$$

```python
stx['Returns'] = np.log( stx.ix['Adj Close'] / stx['Adj Close'].shift(1) )
```

```python
def rolling_ewmcorr_pairwise(df, **kwargs):
    """
    Computes pairwise rolling correlation matrices as Panel whose items are
    dates

    Parameters
    ----------
    df : DataFrame
    window : int
    min_periods : int, default None

    Returns
    -------
    correls : Panel
    """
    from pandas import Panel, ewmcorr
    from collections import defaultdict

    all_results = defaultdict(dict)

    for i, k1 in enumerate(df.columns):
        for k2 in df.columns[i:]:
            corr = ewmcorr(df[k1], df[k2], **kwargs)
            all_results[k1][k2] = corr
            all_results[k2][k1] = corr

    return Panel.from_dict(all_results).swapaxes('items', 'major')
```

Prepare a bokeh columnar data source

```python
names = cmat1.items.values.tolist()
xnames = []
ynames = []
values = []
alphas = []
colors = []
for i, n1 in enumerate(names):
    for i, n2 in enumerate(names):
        xnames.append(n1)
        ynames.append(n2)
        v = cmat1.ix[n1][n2][0]
        values.append(v)
        colors.append("grey")
        alphas.append(np.abs(v))
        
# values = np.abs( cmat1.values.flatten() )
```

```python
from bokeh.plotting import *
from bokeh.objects import HoverTool, ColumnDataSource

source = ColumnDataSource(
    data=dict(
        xname = xnames,
        yname = ynames,
        colors= colors,
        alphas= alphas,
        values= values,
    )
)
```

```python
figure()
rect('xname', 'yname', 0.9, 0.9, source=source,
     x_range=list(reversed(names)), y_range=names,
     color='colors', alpha='alphas', line_color=None,
     tools="resize,hover", title="Correlation matrix",
     plot_width=500, plot_height=500)
```

```python
def get_first_eigenvalues(mat, n_components=5):
    m = np.squeeze(mat.values)
    pca = PCA(n_components=n_components, whiten=False)
    pca.fit(m)
    return pca.explained_variance_
```

```python
from scipy import linalg
def first_eigenvalues(mat, number_values):
    la, v = linalg.eig(np.squeeze(mat))
    return la[:number_values]
```
