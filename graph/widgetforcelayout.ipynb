```python
from IPython.html import widgets
from IPython.display import display
from widgetforcelayout import GraphWidget, publish_js
publish_js()
```

```python
import networkx as nx

mygraph = nx.random_graphs.binomial_graph(20, 0.12, directed=False)
```

```python
g = GraphWidget(width=500, height=500, charge=-250)
display(g)
```

```python
g.set_graph(mygraph)
```
